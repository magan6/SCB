```python
# Exempel: Anropa LLM-tjänsten i SSP Cloud med OpenAI-klienten
# Förutsättning: API-nyckeln finns som miljövariabel via My Secrets (SSP_API_KEY)

from openai import OpenAI
import os

# Hämta API-nyckeln från miljövariabler (injiceras via My Secrets)
api_key = os.environ["SSP_API_KEY"]

# Välj modell manuellt (i Open WebUI kan du se tillgängliga modeller)
model = "gpt-oss:120b"

# Initiera klienten med SSP Clouds LLM-endpoint
client = OpenAI(api_key=api_key,
base_url="https://llm.lab.sspcloud.fr/api")
```

```python
# Skicka ett enkelt testmeddelande
response = client.chat.completions.create(
    model=model,
    messages=[
        {"role": "system", "content": "Du är en hjälpsam assistent som svarar på svenska."},
        {"role": "user", "content": "Beskriv Statistiska centralbyråns kärnverksamhet."}
    ])

# Skriv ut svaret
print(response.choices[0].message.content)
```
